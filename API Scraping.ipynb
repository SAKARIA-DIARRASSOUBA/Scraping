## Format json
cet format va encoder les données préccédentes sous forme de listes pour etre mieux interpreter
**JSON** c'est le format principale pour envoyer et recevoir la data lorqu'on passe par un API
pour lire il faut les librairies:
- librairie json

la méthodes :
- **dumps** prend en entrée un object python et retourne une chaine de caractère
- **loads** prend en entrée une chaine de caractère JSON et retourne un object python(dictionnaire, list,..)

### Exemple

```python
sports=["Tennis","Foot","Triathlon"]
print(type(sports))
```

```python
# importer la librairie json
import json
```

```python
#méthode json.dumps pour convertir en chaine de caractère
sports_string=json.dumps(sports)
print(sports_string)
```

```python
print(type(sports_string))
```

```python
# convertir sports_string en liste-- methode json.loads
sport2=json.loads(sports_string)
print(type(sport2))
```

## Extraire toutes les informations de la page

### Pratique

```python
content=reponse.content
parser=BeautifulSoup(content,'html.parser')
soup=parser.find(id='seven-day-forecast') # trouver le 1er élt id 
forcast_item=soup.find_all(class_='tombstone-container')
seven_day_forcast=[forcast_item[k] for k in range(1,9)]
print(seven_day_forcast)
```

### Correction

```python
seven_day=parser.find(id='seven-day-forecast') # trouver le 1er élt id 

period_tags=seven_day.select(".tombstone-container .period-name") # on applique sur la classe
# tombstone-container et sur la classe period-name
# on parcourt tous les éléments de period_tags pour chaque élément on applique la méthode get_text()
#on obtient une liste
periods=[pt.get_text() for pt in period_tags]
print(periods)
len(periods)
```

```python
temp_tags=seven_day.select(".tombstone-container .temp")
temperature=[tp.get_text() for tp in temp_tags]
print(temperature)
len(temperature)
```

```python
short_tags=seven_day.select(".tombstone-container .short-desc")
descrip=[ tp.get_text() for tp in short_tags]
print(descrip)
len(descrip)
```

```python
desc=[ d['title'] for d in seven_day.select(".tombstone-container img")]
print(desc)
len(desc)
```
