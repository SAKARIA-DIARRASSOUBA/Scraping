## Script pour scraper toutes les pages

## Controler le taux de requetes envoyées

pourquoi besoin de controler le taux de requetes envoyées?
- en web sraping, le risque c'est d'envoyer un trop grand nombres par seconde et que notre IP soit banni
- l'autre risque est de perturber l'utilité du serveur et donc du site web

Donc pour controler ce taux de requetes on utilise la fonction sleep du module time qui va mettre en pause l'execution de notre boucle pour un certain temps

```python
from time import sleep
from random import randint

for a in range(5):
    print('BOOL')
    sleep(randint(1,4))# on affiché BOOL avec une pause de 1 a 4 seconde
```

l'idée de controler le taux de requetes c'est lorsqu'on scrape avec une seule requete des millier de pages

Donc on va écrire un sript qui permet de mesurer la frequence de requetes, le nbre de requetes et aussi retourner le cod_status à savoir reuissite ou echec

```python
from time import time
from IPython.display import clear_output# pour afficher la requete en cours
start=time()# temps du début
requests=0 # la variable requests va compter le nbre de requetes
for k in range(5):
    requests+=1 # on increment la variable requests à chaque passage
    sleep(randint(1,3)) # on fait une pause de 1 à 3 secondes
    end=time()-start # temps écoule pour une requete
    print("requests: {}; frequence : {} requests/s".format(requests,requests/end))  
    clear_output(wait=True)
```

## le script final pour scraper les pages
- redéclarer les listes précédemment crées pour notre script sur une page pour qu'elles deviennent de nouveau vides
- Préparer l'écran d'affichage de la boucle
- Ecrire une boucle qui fait varier le paramètre release_date de l'url avec les valeurs de la liste years_url
- Faire une requete GET sur la boucle des pages
- Pauser la boucle sur un intervalle de 8 à 15 secondes
- Afficher à l'écran le temps de requetes vu à la vidéo précédente
- Ajouter un avertissement pour tout code status différent de 200
- Arreter la boucle si le nbre de requetes est superieur au nbre attendu 72 pages
- Convertir le contenu HTML response en un objet BeautifulSoup
- Ecrire une boucle qui parcourt tout les containers
- Extraire les infos de chaque container si celui-ci a un Metascore

```python
# 1)déclaration des listes vides
Names=[]
Years=[]
Vots=[]
IMDB=[]
MetaScores=[]

# 2) preparer l'ecran l'affichage des boucle
start=time()
requests=0
# 3) une boucle qui fait varier les paramètres
for year_url in years_url:
    # 4) pour chaque page entre 1 et 4
    for page in pages:
        
        # 5) la requete get()
        reponse=get("https://www.imdb.com/search/title/?release_date="+ year_url +"&sort=num_votes,desc&page="+ page )
        
        # 6) pouser la boucle de 8 à 15 secondes
        sleep(randint(8,15))
        
        # 7) afficher le temps de requete
        requests+=1 # on increment la variable requests à chaque passage
        end_url=time()-start
        print("requests:{} ; frequence:{} requete/s".format(requests,requests/end_url))
        clear_output(wait=True)
        
        # 8) ajouter un avertisement si le code status different de 200
        if reponse.status_code!=200:
            warn('request:{}; status code:{}'.format(requests,reponse.status_code))
        
        # 9) verifier la condition d'arret
        if requests >72:
            warn("Nombres de requetes trop important")
            break
        # 10) Convertir le contenu HTML response en un objet BeautifulSoup
        page_html=BeautifulSoup(reponse.text,'html.parser')
        
        # 11) selectionner les 50 films de chaque page (containers)
        films_containers=page_html.find_all("div",class_="lister-item mode-advanced")
        
        # 12) une boucle qui parcourt tout les containers
        for film in films_containers:
            
            # verifions si le film a un metascore
            condit=film.find('div',class_="ratings-metascore")      
            if condit is not None:
                # scraper le score du film
                score=film.find('span',class_="metascore").text
                MetaScores.append(int(score))

                # scraper le nom du film
                nom=film.h3.a.text
                Names.append(nom)

                #scraper l'année de sortie du film
                year=film.h3.find('span',class_="lister-item-year text-muted unbold")
                Years.append(year.text)

                # scraper le nombres de votes
                votes= film.find('span',attrs={'name':"nv"})['data-value']
                Vots.append(int(votes))

                # scraper la note imdb
                imdb=film.strong.text
                IMDB.append(float(imdb))


    
```

```python
movies_ratings=pd.DataFrame({"Film":Names,
                            "Années":Years,
                            "Votes":Vots,
                            "Scores":MetaScores,
                            "IMDB":IMDB})
```
